```python
import subprocess
import os

print("Stopping existing Streamlit processes...")
os.system("pkill -f streamlit")

# Add security bypass flags for JupyterHub Proxy routing
command = [
    "streamlit", "run", "app.py", 
    "--server.port", "8501", 
    "--server.address", "0.0.0.0",
    "--server.enableCORS", "false",
    "--server.enableXsrfProtection", "false"
]

print("Restarting Streamlit with Proxy WebSocket Support...")

# Open text files to capture log dumps
with open("streamlit_stdout.log", "w") as out, open("streamlit_stderr.log", "w") as err:
    process = subprocess.Popen(
        command,
        stdout=out,
        stderr=err,
        preexec_fn=os.setsid
    )

print(f"Streamlit successfully detached! Process ID (PID): {process.pid}")
```
